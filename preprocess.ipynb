```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
import os
from numpy import linalg as LA
from sklearn import preprocessing
```

```python
thai_char = [chr(i)  for i in range(ord('ก'),ord('ฮ')+1)]
col = [f'bin {ch}' for ch in range(1,901)]
col.append('class')
thai_char.remove('ฤ')
thai_char.remove('ฦ')
count = 0
df = pd.DataFrame(columns=col)
```

```python
%%time
hog = cv2.HOGDescriptor((50,50),(5,5),(5,5),(5,5),9)
path  = 'Thai44Consonants/Train/'
for i in range(0,44):
    for j in os.listdir(f'{path}{i}'):
        im = cv2.imread(f'{path}{i}/{j}')
        im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        im = cv2.resize(im, (50, 50))
        im = cv2.GaussianBlur(im, (3, 3), 0)
        h = hog.compute(im)
        dic = {}
        for d in range(0 ,len(h)):
            dic.update({col[d]: h[d]})
        dic.update({'class': thai_char[int(i)]})
        df = pd.concat([df, pd.DataFrame(dic)])
```
